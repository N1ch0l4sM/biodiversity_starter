```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# max number columns to display side by side
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)


observations = pd.read_csv('observations.csv')
species = pd.read_csv('species_info.csv')

species.head()
```

```python
sns.countplot(x='conservation_status', data=species)
plt.xlabel('Conservation Status')
plt.ylabel('Number of Species')
plt.show()
plt.clf()
```

```python
species_endangered = species[species['conservation_status'] == "Endangered"]
sns.countplot(x='category', data=species_endangered)
plt.xlabel('Category')
plt.ylabel('Number of Species')
plt.show()
plt.clf()
```

```python
great_smoky = observations[observations['park_name'] == "Great Smoky Mountains National Park"]
yellowstone = observations[observations['park_name'] == "Yellowstone National Park"]
bryce = observations[observations['park_name'] == "Bryce National Park"]
yosemite = observations[observations['park_name'] == "Yosemite National Park"]

great_smoky_max = great_smoky['observations'].max()
yellowstone_max = yellowstone['observations'].max()
bryce_max = bryce['observations'].max()
yosemite_max = yosemite['observations'].max()

lst = [great_smoky[great_smoky['observations'] == great_smoky_max], yellowstone[yellowstone['observations'] == yellowstone_max] , bryce[bryce['observations'] == bryce_max],yosemite[yosemite['observations'] == yosemite_max]]
park_max = pd.concat(lst)
park_max.head()
```

```python
sns.barplot(x='park_name', y='observations', data=park_max, hue='scientific_name')
plt.xlabel('Scientific Name')
plt.ylabel('Number of Observations')
plt.title('Number of Observations by Scientific Name')
#rotate x-axis labels
plt.xticks(rotation=45)
#change xticks name
plt.xticks(np.arange(0,4), ('Great Smoky Mountains', 'Yellowstone', 'Bryce', 'Yosemite'))
plt.show()
plt.clf()
```
